```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu', input_dim=X_train.shape[1]))
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size=32, epochs=100)
```

for age=30 and estimated salary:30000

```python
customer_info = np.array([[30, 30000]])
customer_info_scaled = sc.transform(customer_info)
prediction = ann.predict(customer_info_scaled)
```

```python
if prediction > 0.5:
    print("The customer will make a purchase.")
else:
    print("The customer will not make a purchase.")
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), axis=1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
matrix = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Confusion Matrix:")
print(matrix)
print(f"Accuracy: {accuracy}")
```
