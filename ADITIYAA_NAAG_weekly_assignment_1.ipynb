# Data Preprocessing

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('SwedishMotorInsurance.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 5:7])
x[:, 5:7] = imputer.transform(x[:, 5:7])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=1)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Linear Regression Model


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('SwedishMotorInsurance.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
prediction = lin_reg.predict([[6, 6, 6, 799, 89]])
print("Linear Regression Prediction:", prediction)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
lin_reg_2.coef_
```

```python
plt.scatter(x[:, 0], y, color='red')
plt.plot(x[:, 0], lin_reg.predict(x), color='blue')
plt.title('Swedish Motor Insurance (Linear Regression)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python

    plt.scatter(x[:, 0], y, color='red')
    plt.plot(x[:, 0], lin_reg_2.predict(poly_reg.fit_transform(x)), color='black')
    plt.title('Swedish Motor Insurance (Polynomial Regression)')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()

```

```python
lin_reg.predict([[2,3,4,111,178]])
```

```python
lin_reg_2.predict(poly_reg.fit_transform([[2,3,4,111,178]]))
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()

x = sc_X.fit_transform(x)
y = y.reshape(-1, 1)
y = sc_y.fit_transform(y)
```

```python
from sklearn.svm import SVR

regressor = SVR(kernel='rbf')
regressor.fit(x, y.ravel())

predictions = regressor.predict(x)
print(predictions)
```

```python
sc_y.inverse_transform(regressor.predict(sc_X.transform([[2,3,4,111,178]])).reshape(-1,1))
```

```python
plt.scatter(x[:, 0], y, color = 'red')
plt.plot(x[:, 0], regressor.predict(x).reshape(-1,1), color = 'blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
